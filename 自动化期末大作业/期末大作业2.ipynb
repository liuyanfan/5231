```python
labels=titan2.embarked.unique().tolist()
labels
```

```python
x=(titan2["sex"]=="male").astype("int")
x
titan2["sex"]=x
```

```python
#titan2["age"].fillna()
titan2["age"].describe()
```

```python
titan2["age"]=titan2["age"].fillna(titan2["age"].mean())
titan2.info()
```

```python
titan2["fare"]=titan2["fare"].fillna(titan2["fare"].mean())
```

```python
titan2.info()
```

```python
titan2.head(3)
```

```python
x=titan2.loc[:,titan2.columns!="survived"]
y=titan2["survived"]
x
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,test_size=0.2)

Xtrain
```

```python
clf=tree.DecisionTreeClassifier()
clf=clf.fit(Xtrain,Ytrain)
score=clf.score(Xtest,Ytest)
score
```

```python
feature_name=Xtrain.columns
feature_name
```

```python
import graphviz
import pydotplus

feature_name_C = ['船舱级别','性别','年龄','子女数','救生艇','票价','登录码头']
dot_data=tree.export_graphviz(clf
                              ,feature_names=feature_name_C
                              ,class_names=["Old","Young","Adult"]
                              ,filled=True
                              ,rounded=True
                                      
                                     )

graph=graphviz.Source(dot_data)
graph

graph1=pydotplus.graph_from_dot_data(dot_data)
graph1.write_pdf("titanic.pdf")
```

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

tr=[]
te=[]

for i in range(10):
    
    clf=tree.DecisionTreeClassifier(random_state=25
                                   ,max_depth=i+1)
    
    clf=clf.fit(Xtrain,Ytrain)
    score_tr=clf.score(Xtrain,Ytrain)
    score_te=cross_val_score(clf,x,y,cv=10).mean()
    tr.append(score_tr)
    te.append(score_te)
    
plt.plot(range(1,11),tr,color="red",label="train")
plt.plot(range(1,11),te,color="blue",label="test")

plt.show()
```
