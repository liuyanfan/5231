```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titan_data=pd.read_csv("titanic3.csv")
titan_data
```

```python
#不同性别对生还率的影响
df_sex1=titan_data['sex'][titan_data['survived']==1]
df_sex0=titan_data['sex'][titan_data['survived']==0]
plt.hist([df_sex1,df_sex0],
stacked=True,
label=['Rescued','not saved'])
plt.xticks([0,1],['female','male'])
plt.legend()
plt.title('Sex_Survived')
```

```python
#不同等级对生还率的影响
df_pc1=titan_data['pclass'][titan_data['survived']==1]
df_pc0=titan_data['pclass'][titan_data['survived']==0]
plt.hist([df_pc1,df_pc0],
stacked=True,
label=['Rescued','not saved'])
plt.xticks([1,2,3],['Upper','Middle','lower'])
plt.legend()
plt.title('Pclass_Survived')
```

```python
#不同年龄对生还率的影响
df_pc1=titan_data['age'][titan_data['survived']==1]
df_pc0=titan_data['age'][titan_data['survived']==0]
plt.hist([df_pc1,df_pc0],
stacked=True,
label=['Rescued','not saved'])
plt.legend()
plt.title('Age_Survived')
```

```python
# 年龄，阶层双因素
fig, ax = plt.subplots(1, 3,  figsize = (18, 8))
sns.violinplot("pclass", "age", hue="survived", data=titan_data, split=True, ax=ax[0])
ax[0].set_title('Pclass and Age vs Survived')
ax[0].set_yticks(range(0, 110, 10))
 # 性别，年龄双因素
sns.violinplot("sex", "age", hue="survived", data=titan_data, split=True, ax=ax[1])
ax[1].set_title('Sex and Age vs Survived')
ax[1].set_yticks(range(0, 110, 10))
#性别，阶层
sns.violinplot("sex", "pclass", hue="survived", data=titan_data, split=True, ax=ax[2])
ax[2].set_title('Sex and Pclass vs Survived')
```

```python
# 查看相关性
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(titan_data.astype(float).corr(),linewidths=0.2,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```
